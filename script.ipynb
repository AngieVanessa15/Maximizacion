### Modelo PML 

from pulp import LpMaximize, LpProblem, LpVariable

# Datos de ejemplo (Prospectos cambiarlos por concesionarios y su porcentaje de eficiencia)
prospectos = {
    'prospecto1': {'eficiencia': 0.8},
    'prospecto2': {'eficiencia': 0.6},
    'prospecto3': {'eficiencia': 0.7},
    # Agregar más prospectos con sus respectivos porcentajes de eficiencia
}

limite_prospectos = 10  # Límite de prospectos a distribuir

# Crear el problema de optimización
problema = LpProblem("MaximizarDistribucionProspectos", LpMaximize)

# Variables de decisión: si se distribuye o no cada prospecto
distribuir = LpVariable.dicts("Distribuir", prospectos.keys(), cat='Binary')

# Función objetivo: maximizar la eficiencia total
problema += sum(prospectos[prospecto]['eficiencia'] * distribuir[prospecto] for prospecto in prospectos.keys()), "EficienciaTotal"

# Restricción: número máximo de prospectos a distribuir
problema += sum(distribuir[prospecto] for prospecto in prospectos.keys()) <= limite_prospectos, "LimiteProspectos"

# Resolver el problema
problema.solve()

# Mostrar el resultado
print("Estado de la solución:", problema.status)
print("Cantidad total de prospectos distribuidos:", sum(distribuir[prospecto].varValue for prospecto in prospectos.keys()))

# Mostrar los prospectos distribuidos
for prospecto in prospectos.keys():
    if distribuir[prospecto].varValue == 1:
        print("Distribuir", prospecto)
